```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
```

```python
iv = .55
s0 = 100
dte = 30
```

IV = expected 1 standard deviation move over 1 year
1. Rescale to 1y
2. Compute estimated distribution (1sd, 2sd, ...)

```python
dte = 30
dte_range = np.linspace(start=1, stop=dte, num=dte)

# DOUBLE CHECK THIS FORMULA
iv_scaled = iv * np.sqrt(dte_range/365)
sd = iv_scaled*s0

x_range = np.linspace(s0*.5, s0*1.5, num=200)
df_dist = pd.DataFrame(
    np.array([stats.norm.pdf(x=x_range, loc=s0, scale=s) for s in sd]).T
)
plt.plot(df_dist)
plt.xlabel('Underlying Price')
plt.ylabel('Probability')
plt.show()
```

```python
r = 0.05
t = dte/365
exp_1sd_move = s0*np.exp(r*t+iv*np.sqrt(t))
print(exp_1sd_move)
```
