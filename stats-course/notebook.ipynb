```python
import numpy as np
from scipy.stats import norm
import seaborn as sns
```

```python
mu = 1
sigma = 2
```

```python
# P(X <= 1)
norm.cdf(
    x=1,
    loc=mu,
    scale=sigma
)
```

```python
# P(X >= -2)
1 - norm.cdf(-2,mu,sigma)
```

```python
X_dist = norm.rvs(loc=mu, scale=sigma, size=1000000)
sns.histplot(X_dist)
```

```python
(X_dist >= -2).sum()/len(X_dist)
```
